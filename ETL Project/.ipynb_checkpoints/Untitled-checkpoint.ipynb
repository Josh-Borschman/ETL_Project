```python
import pandas as pd
import glob
```

```python
files = sorted(glob.glob("Resources/*.csv"))
dfs = [pd.read_csv(f, encoding= 'ISO-8859-1') for f in files]
```

```python
df2015 = dfs[0]
df2015.head()
df2015 = df2015.drop(['Region', 'Standard Error', 'Family', 'Freedom', 'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual'], axis=1)
df2015.rename(columns={"Happiness Rank": "Happiness_Rank", "Happiness Score": "Happiness_Score", "Economy (GDP per Capita)":"GDP_Per_Capita", "Health (Life Expectancy)": "Health(Life_Expectancy)"}, inplace=True)
df2015["Year"] = '2015'
df2015.set_index('Country', inplace=True)
df2015.head()
```

```python
df2016 = dfs[1]
df2016.head()
df2016 = df2016.drop(['Region', 'Lower Confidence Interval', 'Upper Confidence Interval', 'Family', 'Freedom', 'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual'], axis=1)
df2016.rename(columns={"Happiness Rank": "Happiness_Rank", "Happiness Score": "Happiness_Score", "Economy (GDP per Capita)":"GDP_Per_Capita", "Health (Life Expectancy)": "Health(Life_Expectancy)"}, inplace=True)
df2016["Year"] = '2016'
df2016.set_index('Country', inplace=True)
df2016.head()
```

```python
df2017 = dfs[2]
df2017.head()
df2017 = df2017.drop(['Whisker.high', 'Whisker.low', 'Family', 'Freedom', 'Trust..Government.Corruption.', 'Generosity', 'Dystopia.Residual'], axis=1)
df2017.rename(columns={"Happiness.Rank": "Happiness_Rank", "Happiness.Score": "Happiness_Score", "Economy..GDP.per.Capita.": "GDP_Per_Capita", "Health..Life.Expectancy.": "Health(Life_Expectancy)"}, inplace=True)
df2017["Year"] = '2017'
df2017.set_index('Country', inplace=True)
df2017.head()
```

```python
df2018 = dfs[3]
df2018.head()
df2018 = df2018.drop(['Social support', 'Freedom to make life choices', 'Generosity','Perceptions of corruption'], axis=1)
df2018.rename(columns={"Overall rank": "Happiness_Rank", "Country or region": "Country", "Score": "Happiness_Score", "GDP per capita" :"GDP_Per_Capita", "Healthy life expectancy": "Health(Life_Expectancy)"}, inplace=True)
df2018["Year"] = '2018'
df2018.set_index('Country', inplace=True)
df2018.head()
```

```python
df2019 = dfs[4]
df2019.head()
df2019 = df2019.drop(['Social support', 'Freedom to make life choices', 'Generosity','Perceptions of corruption'], axis=1)
df2019.rename(columns={"Overall rank": "Happiness_Rank", "Country or region": "Country", "Score": "Happiness_Score", "GDP per capita" :"GDP_Per_Capita", "Healthy life expectancy": "Health(Life_Expectancy)"}, inplace=True)
df2019["Year"] = '2019'
df2019.set_index('Country', inplace=True)
df2019.head()
```

```python
combined_df = pd.concat([df2015,df2016,df2017, df2018, df2019],axis=0,sort=False)
combined_df
```

```python
list_of_countries = combined_df.index.to_list()
list1 = set(list_of_countries)
list1
```
