# Corelation between Infant Mortality Rate and Happiness Score

```python
# Displaying the happiness dataframe
happiness
```

```python
# Displaying the mortality dataframe
mortality1 = mortality.loc[(mortality['year'] >=2015) & (mortality ['gender'] =='Total') ]
mortality1
```

```python
# Merging the dataframes on country id and year
merged_df = pd.merge(happiness, mortality1,  how='left', on=['country_id', 'year'])
```

```python
# Dropping any rows with NAs
merged_df = merged_df.dropna(axis=0, how='any')
```

```python
# Plot out happiness score versus infant mortality rate
x_values = merged_df['happiness_score']
y_values = merged_df['infant_mortality_rate']
# Getting all the statistical values
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
# Plotting the graph and setting x label and y label
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Happiness Rank')
plt.ylabel('Infant Mortality Rate')
# Print out the r-squared value along with the plot.
print(f"The r-squared is: {rvalue**2}")
plt.show()
```
