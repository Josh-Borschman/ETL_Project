```python
import pandas as pd
import glob
from sqlalchemy import create_engine
```

```python
files = sorted(glob.glob("Resources/*.csv"))
dfs = [pd.read_csv(f, encoding= 'ISO-8859-1') for f in files]
```

```python
df2015 = dfs[0]
df2015.head()
df2015 = df2015.drop(['Region', 'Standard Error', 'Family', 'Freedom', 'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual'], axis=1)
df2015.rename(columns={ "Country": "country","Happiness Rank": "happiness_rank", "Happiness Score": "happiness_score", "Economy (GDP per Capita)":"gdp_per_capita", "Health (Life Expectancy)": "health_life_expectancy"}, inplace=True)
df2015["year"] = '2015'
df2015.set_index('country', inplace=True)
df2015.head()
```

```python
df2016 = dfs[1]
df2016.head()
df2016 = df2016.drop(['Region', 'Lower Confidence Interval', 'Upper Confidence Interval', 'Family', 'Freedom', 'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual'], axis=1)
df2016.rename(columns={"Country": "country","Happiness Rank": "happiness_rank", "Happiness Score": "happiness_score", "Economy (GDP per Capita)":"gdp_per_capita", "Health (Life Expectancy)": "health_life_expectancy", }, inplace=True)
df2016["year"] = '2016'
df2016.set_index('country', inplace=True)
df2016.head()
```

```python
df2017 = dfs[2]
df2017.head()
df2017 = df2017.drop(['Whisker.high', 'Whisker.low', 'Family', 'Freedom', 'Trust..Government.Corruption.', 'Generosity', 'Dystopia.Residual'], axis=1)
df2017.rename(columns={"Country": "country","Happiness.Rank": "happiness_rank", "Happiness.Score": "happiness_score", "Economy..GDP.per.Capita.": "gdp_per_capita", "Health..Life.Expectancy.": "health_life_expectancy"}, inplace=True)
df2017["year"] = '2017'
df2017.set_index('country', inplace=True)
df2017.head()
```

```python
df2018 = dfs[3]
df2018.head()
df2018 = df2018.drop(['Social support', 'Freedom to make life choices', 'Generosity','Perceptions of corruption'], axis=1)
df2018.rename(columns={"Country": "country","Overall rank": "happiness_rank", "Country or region": "country", "Score": "happiness_score", "GDP per capita" :"gdp_per_capita", "Healthy life expectancy": "health_life_expectancy"}, inplace=True)
df2018["year"] = '2018'
df2018.set_index('country', inplace=True)
df2018.head()
```

```python
df2019 = dfs[4]
df2019.head()
df2019 = df2019.drop(['Social support', 'Freedom to make life choices', 'Generosity','Perceptions of corruption'], axis=1)
df2019.rename(columns={"Country": "country","Overall rank": "happiness_rank", "Country or region": "country", "Score": "happiness_score", "GDP per capita" :"gdp_per_capita", "Healthy life expectancy": "health_life_expectancy"}, inplace=True)
df2019["year"] = '2019'
df2019.set_index('country', inplace=True)
df2019.head()
```

```python
combined_df = pd.concat([df2015,df2016,df2017, df2018, df2019],axis=0,sort=False)
combined_df1 = combined_df.copy()
combined_df1 = combined_df1.reset_index()
combined_df1 = combined_df1.drop(["country"], axis=1)
combined_df1
```

```python
infant_df = pd.read_csv("Rate/InfantMortalityRate.csv", encoding= 'ISO-8859-1')
infant_df = infant_df.dropna()
infant_df1 = infant_df.copy()
infant_df1['Year'] = infant_df.Year.astype(int)
infant_df1 = infant_df1.rename(columns={"Country": "country","Infant Mortality Rate": "infant_mortality_rate", "Gender":"gender", "Year": "year"}, inplace=False)
del infant_df1['country']
```
